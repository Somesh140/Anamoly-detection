```python
with open("data0 (1).txt","r") as f:
    df1=f.read()
    df2=df1.strip().split(",")
df2.pop()
print(df2)
```

```python
df3=pd.DataFrame(df2,columns=["value"])
df3['value']=df3['value'].astype('int')
df3
```

```python
from sklearn.ensemble import IsolationForest
```

```python
anomaly_inputs = ["value"]
model_IF = IsolationForest(contamination=float(0.1),random_state=42)
model_IF.fit(df3[anomaly_inputs])
df3['anomaly_scores'] = model_IF.decision_function(df3[anomaly_inputs])
df3['anomaly'] = model_IF.predict(df3[anomaly_inputs])
```
